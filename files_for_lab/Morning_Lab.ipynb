```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
file = pd.read_csv("marketing_customer_analysis.csv")
file.head()
```

```python
# Show DataFrame info.
file.info()
```

```python
# Describe DataFrame.
file.describe(exclude=[object])
```

```python
# Show a plot of the total number of responses.
file[file.Response != "No"]
counts = file.Response.value_counts()
ax = sns.barplot(x = file.Response.unique(), y = counts, data = file)
```

```python
# Show a plot of the response rate by the sales channel.
sns.countplot(x = "Sales Channel", hue = "Response", data = file)
plt.show()
```

```python
# Show a plot of the response rate by the total claim amount
sns.displot(file, x="Total Claim Amount", hue = "Response", bins = 20)
plt.show()
```

```python
# Show a plot of the response rate by income.
sns.displot(file, x="Income", hue = "Response", bins = 20)
plt.show()
```
